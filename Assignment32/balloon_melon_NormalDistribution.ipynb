```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balloon_width=np.random.normal(50,5,50)
balloon_length=np.random.normal(50,5,50)
balloon_weight=np.random.normal(0.002,0.001,50)
                                
melon_width=np.random.normal(15,3,50)
melon_length=np.random.normal(20,3,50)
melon_weight=np.random.normal(6,1,50)
                              
    
                                
                                
```

```python
melon_width
```

```python
width=np.concatenate((balloon_width,melon_width))
length=np.concatenate((balloon_length,melon_length))
weight=np.concatenate((balloon_weight,melon_weight))
data=np.array([width,length,weight]).T
data.shape
```

```python
fig=plt.figure()
p=plt.axes(projection='3d')
#balloons are blue
#melons are orange
#weight in kilos
p.scatter(data[0:50,0],data[0:50,1],data[0:50,2],c='blue',marker='^',label='Balloons')
p.scatter(data[50:,0],data[50:,1],data[50:,2],c='orange',label='Melons')
p.set_xlabel('Width')
p.set_ylabel('Length')
p.set_zlabel('Weight')
plt.title('Normal distribution of balloons and melons features')
plt.legend()
plt.savefig('Normal distribution of balloons and melons features.png')
```
