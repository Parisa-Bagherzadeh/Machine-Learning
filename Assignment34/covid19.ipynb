```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid=pd.read_csv('/COVID.csv')
covid.head()
```

```python
covid['Date']=pd.to_datetime(covid['Date'])
covid['Date']=covid['Date'].map(lambda  Date: 100*Date.year + Date.month)
covid.head()
```

```python
country=covid.query('Date==202004')
country
```

```python
country=country.groupby(['Country/Region'])['Confirmed'].sum().reset_index()
country
```

```python
country=country.sort_values(by="Confirmed",ascending=False).reset_index()
country.head(8)
```
