```python
import random
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv
from scipy import stats
```

```python
Study_time=[]
Grade=[]
N=50

for i in range(N):
    study_time=random.randint(0,15)
    Study_time.append(study_time)
    grade=study_time+random.randint(0,20)*0.2
    Grade.append(grade)




Study_time=np.array(Study_time)
Grade=np.array(Grade)
Study_time=Study_time.reshape(N,1)
Grade=Grade.reshape(N,1)

```

```python
plt.scatter(Study_time,Grade,c='blue')
plt.xlabel('Study Time')
plt.ylabel('Grade')
plt.title('Random data')
```

```python
#random slope

plt.scatter(Study_time,Grade,c='blue')

for i in range(5):
    m=random.random()*1.5
    y=np.multiply(Study_time,m)
    plt.plot(Study_time,y,c='hotpink')


plt.xlabel('Study Time')
plt.ylabel('Grade')
plt.title('Random lines on data')
    
    
```

```python
#m
m=np.matmul(inv(np.matmul(Study_time.T,Study_time)),np.matmul(Study_time.T,Grade))
```

```python
Y_pred=np.matmul(Study_time,m)
```

```python
plt.scatter(Study_time,Grade,c='blue',label='data')
plt.plot(Study_time,Y_pred,c='red',label='Linear Regression')
line,intercept,_,_,_=stats.linregress(Study_time[:,0],Grade[:,0])
plt.plot(Study_time,intercept+line*Study_time,'g',label='Linregress Method')
plt.xlabel('Study Time')
plt.ylabel('Grade')
plt.legend()
plt.title('Compare fitted lines')
```
